#### 2. ë°ì´í„° ì „ì²˜ë¦¬

```python
# ë°ì´í„°, ë¼ë²¨ ì§€ì •
wine_data = wine.data
wine_label = wine.target
```

```python
# train, test ë°ì´í„° ë¶„ë¦¬
X_train, X_test, y_train, y_test = train_test_split(wine_data,
                                                    wine_label,
                                                    test_size=0.2,
                                                    random_state=1)
print("X_train ê°¯ìˆ˜:", len(X_train))
print("X_test ê°¯ìˆ˜:", len(X_test))
```

#### 3. ëª¨ë¸ í•™ìŠµ ë° ì˜ˆì¸¡

- **Decision Tree:** ê³„ì‚°ë³µì¡ì„± ëŒ€ë¹„ ë†’ì€ ì˜ˆì¸¡ ì„±ëŠ¥ì„ ê°–ê³  ìˆì§€ë§Œ, ê²°ì •ê²½ê³„(decision boundary)ê°€ ë°ì´í„° ì¶•ì— ìˆ˜ì§ì´ì–´ì„œ íŠ¹ì • ë°ì´í„°ì—ë§Œ ì˜ ì‘ë™í•  ê°€ëŠ¥ì„±ì´ ë†’ìŒ
- **Random Forest:** ì˜ì‚¬ê²°ì •ë‚˜ë¬´ë¥¼ ì—¬ëŸ¬ ê°œ ë§Œë“¤ì–´ ê·¸ ê²°ê³¼ë¥¼ ì¢…í•©í•´ ì˜ˆì¸¡
- **SVM:** ë°ì´í„°ë¥¼ ë¹„ì„ í˜• ë§¤í•‘(Mapping)ì„ í†µí•´ ê³ ì°¨ì›ìœ¼ë¡œ ë³€í™˜í•˜ê³ , ìƒˆë¡œìš´ ì°¨ì›ì—ì„œ ì´ˆí‰ë©´(hyperplane)ì„ ìµœì ìœ¼ë¡œ ë¶„ë¦¬í•˜ëŠ” ì„ í˜•ë¶„ë¦¬ë¥¼ ì°¾ì•„ì„œ ìµœì ì˜ ì˜ì‚¬ê²°ì • ì˜ì—­(Decision Boundary)ë¥¼ ì°¾ìŒ
- **SGD Classifier:** í™•ë¥ ì ìœ¼ë¡œ ë¬´ì‘ìœ„ë¡œ ê³¨ë¼ë‚¸ ë°ì´í„°ì— ëŒ€í•´ ìˆ˜í–‰í•˜ëŠ” ê²½ì‚¬ í•˜ê°•ë²•
- **Logistic Regession:** íšŒê·€ë¥¼ ì‚¬ìš©í•˜ì—¬ ë°ì´í„°ê°€ ì–´ë–¤ ë²”ì£¼ì— ì†í•  í™•ë¥ ì„ 0ì—ì„œ 1 ì‚¬ì´ì˜ ê°’ìœ¼ë¡œ ì˜ˆì¸¡í•˜ê³ , ê·¸ í™•ë¥ ì— ë”°ë¼ ê°€ëŠ¥ì„±ì´ ë” ë†’ì€ ë²”ì£¼ì— ì†í•˜ëŠ” ê²ƒìœ¼ë¡œ ë¶„ë¥˜

```python
# Decision Tree
decision_tree = DecisionTreeClassifier(random_state=2)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# Random Forest
random_forest = RandomForestClassifier(random_state=2)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# SVM
svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# SGD Classifier
sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# Logistic Regression
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print(classification_report(y_test, y_pred))
```
